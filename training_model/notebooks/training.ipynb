```python
# GPU + version checks

import torch, ultralytics

print("PyTorch:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
print("Device:", torch.cuda.get_device_name() if torch.cuda.is_available() else "CPU")
print("Ultralytics:", ultralytics.__version__)
```

```python
# model training

from ultralytics import YOLO

model = YOLO("yolov8m.pt")

model.train(
    data="../dataset.yaml",
    epochs=50,
    imgsz=1024,
    batch=8,
    workers=4,
    project="runs/detect",
    name="train",
    exist_ok=True
)
```

```python
# test trained model (make sure to put the best.pt file in "models/trained") 

from ultralytics import YOLO

model = YOLO("../models/trained/best.pt")

model.predict(
    source="../data/images/predict/predictfloor.png", # test image name
    project="runs/detect",
    name="predict",
    conf=0.25,
    save=True
)
```
